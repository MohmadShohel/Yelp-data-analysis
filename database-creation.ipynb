```python
CSV_PATH = "/kaggle/working/user_data.csv"
chunk = 3000
engine = create_engine('mysql+mysqlconnector://admin:qhwtUa63wETqh8c9Lnfx@temp-db.clqg0mkouhbg.ap-south-1.rds.amazonaws.com:3306/datascience')
chunks_count = sum(1 for row in open('/kaggle/working/user_data.csv', 'r'))

def process_csv_in_chunks(csv_file, chunk_size):
    # Read the CSV file in chunks
    csv_chunks = pd.read_csv(csv_file, chunksize=chunk_size)
    curr = 0
    for chunk in csv_chunks:
        # Insert each chunk into the MySQL table
        curr= curr +1
        chunk.to_sql('user', con=engine, if_exists='append', index=False)
        print(f"Inserted chunk of size {len(chunk)} ",end="   ")
        print("Inserting: ", curr, " OF ", chunks_count/chunk_size)

# Call the function to process and insert the CSV in chunks
process_csv_in_chunks(CSV_PATH, chunk)
```
